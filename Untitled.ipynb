```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
```

```python
def get_mean(data, column_name):
    return data[column_name + '_x'] + data[column_name + '_y'] + data[column_name]

def get_sd(data, mean, column_name):
    sd = mean // mean
    for i, x in enumerate([column_name + '_x', column_name + '_y', column_name]) :
        sd += (data[x] - mean)**2
    sd /= 3
    return sd.apply(lambda x : x**(.5))

def smooth(steps, values, steps_new):
    spl = make_interp_spline(steps, values, k=1) #BSpline object
    return spl(steps_new)
```

```python
# create csv files (must be 3 sets of data)
fileName0 = '1D_advanced_Sequential1000_EpsGreedyQ_10000steps(0)'
fileName1 = '1D_advanced_Sequential1000_EpsGreedyQ_10000steps(1)'
fileName2 = '1D_advanced_Sequential1000_EpsGreedyQ_10000steps(2)'

file0 = pd.read_csv(inputPath + fileName0 + '.csv')
file1 = pd.read_csv(inputPath + fileName1 + '.csv')
file2 = pd.read_csv(inputPath + fileName2 + '.csv')
```

```python
# merge data of the 3 sets
data = file1.merge(file0, on = 'step').merge(file2, on = 'step')
```

```python
metric_name = 'loss'
```

```python
# calculate mean and sd on data
mean = get_mean(data, metric_name)
sd = get_sd(data, mean, metric_name)
min_values = mean - 2 * sd/(3)**(.5)
max_values = mean + 2 * sd/(3)**(.5)
```

```python
steps = data['step']
```

```python
plt.plot(steps, min_values)
plt.plot(steps, max_values)
plt.xlabel('step')
plt.ylabel(metric_name)
plt.title(fileName0 + ' ' + metric_name)
plt.legend()
plt.show()
```
